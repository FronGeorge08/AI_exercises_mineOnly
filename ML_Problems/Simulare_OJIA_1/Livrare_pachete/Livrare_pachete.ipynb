```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
```

```python
train=pd.read_csv("/content/train_data.csv")
test=pd.read_csv("/content/test_data.csv")
```

```python
subtask1=round(test["traffic_level"].mean(),2)
subtask2=round(test["traffic_level"].std(),2)
```

```python
features=["distance_km","package_weight_kg","traffic_level"]
x_train=train[features]
x_test=test[features]
y_train=train["on_time"]
```

```python
preprocessor=ColumnTransformer(transformers=[
    ('numeric',StandardScaler(),features)
])
```

```python
model=Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('regressor',RandomForestClassifier(
     n_estimators=300,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features='sqrt',
    n_jobs=-1,
    random_state=42))
])
```

```python
model.fit(x_train,y_train)
y_test=model.predict(test[features])
print (y_test)
```

```python
solution=pd.DataFrame({
    'subtaskID':[1,2],
    'datapointID':[1,1],
    'answer':[subtask1,subtask2]
})
for i in range(test.shape[0]):
    id=test["id"][i]
    sol=pd.DataFrame({
        'subtaskID':[3],
        'datapointID':[id],
        'answer':[y_test[i]]
    })
    solution=pd.concat([solution,sol],ignore_index=True)
solution.to_csv("submission.csv",index=False)
```
